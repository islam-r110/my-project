# Polynomial Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('housing.csv')
X = dataset[['RM']]
y = dataset['MEDV']
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Linear Regression model on the training set

```
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

## Training the Polynomial Regression model on the training set

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X_train)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y_train)
```

## Visualising the Linear Regression results

```
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, lin_reg.predict(X_train), color = 'blue')
plt.title('Linear Regression')
plt.xlabel('Average numper of rooms(RM)')
plt.ylabel('Median value of home (MEDV)')
plt.show()
```

## Visualising the Polynomial Regression results

```
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, lin_reg_2.predict(poly_reg.fit_transform(X_train)), color = 'blue')
plt.title('Polynomial Regression')
plt.xlabel('Average numper of rooms(RM)')
plt.ylabel('Median value of home (MEDV)')
plt.show()
```

## Predicting a new result with Linear Regression

```
lin_reg.predict([[6.998]])
```

## Predicting a new result with Polynomial Regression

```
lin_reg_2.predict(poly_reg.fit_transform([[6.998]]))
```
